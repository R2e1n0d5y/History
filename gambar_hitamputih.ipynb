```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
image = cv.imread('beringin.png', cv.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
print(image.shape)
```

```python
def thresholding(citra, ambang):    
    height, width = citra.shape
    hasil = np.zeros((height, width))

    for i in range(height):
        for j in range(width):
            if citra[i, j] >= ambang:
                hasil[i, j] = 255
            else:
                hasil[i, j] = 0

    return hasil
```

```python
thres_img = thresholding(image, 100)
thres_img = 255 - thres_img
plt.imshow(thres_img, cmap='gray')
print(thres_img.shape)
```

```python
cv.imwrite('pohon_hitamputih.png', thres_img)
```
